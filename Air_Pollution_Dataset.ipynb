```python
# Define bins and labels for categorizing pollutants
aqi_bins = [0, 50, 100, 150, 200, 300, 500]
aqi_labels = ['Good', 'Moderate', 'Unhealthy for Sensitive Groups', 'Unhealthy', 'Very Unhealthy', 'Hazardous']

pm25_bins = [0, 50, 100, 150, 200, 300, 500]
pm25_labels = ['Good', 'Moderate', 'Unhealthy for Sensitive Groups', 'Unhealthy', 'Very Unhealthy', 'Hazardous']

pm10_bins = [0, 50, 100, 150, 200, 300, 500]
pm10_labels = ['Good', 'Moderate', 'Unhealthy for Sensitive Groups', 'Unhealthy', 'Very Unhealthy', 'Hazardous']

o3_bins = [0, 50, 100, 150, 200, 300, 500]
o3_labels = ['Good', 'Moderate', 'Unhealthy for Sensitive Groups', 'Unhealthy', 'Very Unhealthy', 'Hazardous']

co_bins = [0, 1, 2, 10, 17, 34, 50]
co_labels = ['Good', 'Moderate', 'Unhealthy for Sensitive Groups', 'Unhealthy', 'Very Unhealthy', 'Hazardous']

so2_bins = [0, 20, 80, 250, 500, 750, 1000]
so2_labels = ['Good', 'Moderate', 'Unhealthy for Sensitive Groups', 'Unhealthy', 'Very Unhealthy', 'Hazardous']

no2_bins = [0, 40, 80, 180, 280, 400, 500]
no2_labels = ['Good', 'Moderate', 'Unhealthy for Sensitive Groups', 'Unhealthy', 'Very Unhealthy', 'Hazardous']

# Apply categorization
df['AQI_Class'] = pd.cut(df['AQI'], bins=aqi_bins, labels=aqi_labels)
df['PM2.5_Class'] = pd.cut(df['PM2.5'], bins=pm25_bins, labels=pm25_labels)
df['PM10_Class'] = pd.cut(df['PM10'], bins=pm10_bins, labels=pm10_labels)
df['O3_Class'] = pd.cut(df['O3'], bins=o3_bins, labels=o3_labels)
df['CO_Class'] = pd.cut(df['CO'], bins=co_bins, labels=co_labels)
df['SO2_Class'] = pd.cut(df['SO2'], bins=so2_bins, labels=so2_labels)
df['NO2_Class'] = pd.cut(df['NO2'], bins=no2_bins, labels=no2_labels)

# Display value counts for each class
print("AQI Class Counts:")
print(df['AQI_Class'].value_counts())

print("\nPM2.5 Class Counts:")
print(df['PM2.5_Class'].value_counts())

print("\nPM10 Class Counts:")
print(df['PM10_Class'].value_counts())

print("\nO3 Class Counts:")
print(df['O3_Class'].value_counts())

print("\nCO Class Counts:")
print(df['CO_Class'].value_counts())

print("\nSO2 Class Counts:")
print(df['SO2_Class'].value_counts())

print("\nNO2 Class Counts:")
print(df['NO2_Class'].value_counts())
```

```python
# Calculating the range for each pollutant
range_values = {
    "AQI": (df['AQI'].min(), df['AQI'].max()),
    "PM2.5": (df['PM2.5'].min(), df['PM2.5'].max()),
    "PM10": (df['PM10'].min(), df['PM10'].max()),
    "O3": (df['O3'].min(), df['O3'].max()),
    "CO": (df['CO'].min(), df['CO'].max()),
    "SO2": (df['SO2'].min(), df['SO2'].max()),
    "NO2": (df['NO2'].min(), df['NO2'].max())
}

# Displaying the range for each pollutant
print("Range of Values:")
for pollutant, value_range in range_values.items():
    print(f"{pollutant}: {value_range}")
```

```python
# Calculating the range for each pollutant
range_AQI = df['AQI'].max() - df['AQI'].min()
range_PM25 = df['PM2.5'].max() - df['PM2.5'].min()
range_PM10 = df['PM10'].max() - df['PM10'].min()
range_O3 = df['O3'].max() - df['O3'].min()
range_CO = df['CO'].max() - df['CO'].min()
range_SO2 = df['SO2'].max() - df['SO2'].min()
range_NO2 = df['NO2'].max() - df['NO2'].min()

# Displaying the range for each pollutant
print("The range of AQI is {}".format(range_AQI))
print("The range of PM2.5 is {}".format(range_PM25))
print("The range of PM10 is {}".format(range_PM10))
print("The range of O3 is {}".format(range_O3))
print("The range of CO is {}".format(range_CO))
print("The range of SO2 is {}".format(range_SO2))
print("The range of NO2 is {}".format(range_NO2))
```

## **Exploratory Analysis and Visualization**

```python
# Creating a colorful histogram for the distribution of AQI values
plt.figure(figsize=(10, 8))
sns.histplot(df['AQI'], bins=50, kde=True, lw=2, kde_kws={'bw_adjust': 0.1})
plt.title('Distribution of AQI Values')
plt.xlabel('AQI')
plt.ylabel('Frequency')
plt.show()

# Defining a custom order for AQI categories
aqi_order = ['Good', 'Moderate', 'Unhealthy for Sensitive Groups', 'Unhealthy', 'Very Unhealthy', 'Severe']

# Creating a colorful bar chart with updated syntax to avoid the FutureWarning
plt.figure(figsize=(10, 8))
sns.countplot(data=df, x='AQI_Class', hue='AQI_Class', palette='viridis', order=aqi_order, dodge=False)
plt.legend([],[], frameon=False)
plt.title('Distribution of AQI Categories')
plt.xlabel('AQI Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Calculating the mean AQI for each location and sorting in descending order
mean_aqi_per_location = df.groupby('Location')['AQI'].mean().sort_values(ascending=False)

# Selecting the top 10 locations with the highest mean AQI values
top_10_locations = mean_aqi_per_location.head(10)

# Filtering the dataset to include only data from the top 10 locations
top_10_data = df[df['Location'].isin(top_10_locations.index)]

# Creating a table to display mean AQI values for the top 10 locations
top_10_table = top_10_locations.reset_index()
top_10_table.columns = ['Location', 'Mean AQI']

# Displaying the table
top_10_table
```

```python
# Calculating the mean AQI for each location and selecting the top 10
mean_aqi_per_location = df.groupby('Location')['AQI'].mean().sort_values(ascending=False)
top_10_locations = mean_aqi_per_location.head(10)
top_10_table = top_10_locations.reset_index()
top_10_table.columns = ['Location', 'Mean AQI']

# Creating a bar chart to compare mean AQI values across the top 10 locations with updated syntax
plt.figure(figsize=(12, 8))
sns.barplot(data=top_10_table, x='Location', y='Mean AQI', hue='Location', palette='viridis', dodge=False)
plt.legend([],[], frameon=False)
plt.title('Top 10 Locations with Highest Mean AQI Values')
plt.xlabel('Location')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Mean AQI')
plt.show()
```

```python
# Selecting relevant columns for correlation analysis
relevant_columns = ['AQI', 'PM2.5', 'PM10', 'O3', 'CO', 'SO2', 'NO2']
correlation_matrix = df[relevant_columns].corr()

# Creating a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Air Quality Indicators')
plt.show()
```

```python
# Selecting relevant columns for the pairplot
relevant_columns = ['AQI', 'PM2.5', 'PM10', 'O3', 'CO', 'SO2', 'NO2']

# Creating a pairplot to visualize pairwise relationships
sns.pairplot(df[relevant_columns], diag_kind='kde', plot_kws={'alpha':0.5})
plt.suptitle('Pairwise Relationships Between Air Quality Metrics', y=1.02)  # Adjust title position
plt.show()
```

```python
# Calculating the mean values for each pollutant
mean_pollutants = df[['PM2.5', 'PM10', 'O3', 'CO', 'SO2', 'NO2']].mean()

# Calculating pollutant contributions as a percentage of the overall mean AQI
overall_mean_aqi = df['AQI'].mean()
pollutant_contributions = (mean_pollutants / overall_mean_aqi) * 100

# Plotting the pie chart for individual pollutant contributions as a percentage of AQI
plt.figure(figsize=(8, 8))
plt.pie(pollutant_contributions, labels=pollutant_contributions.index, autopct='%1.1f%%', startangle=140)
plt.title('Individual Pollutant Contributions as a Percentage of AQI')
plt.axis('equal')
plt.show()
```
